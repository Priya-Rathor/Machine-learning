```python
import requests
import pandas as pd

all_dfs = []

for j in range(1, 501):  
    url = f"https://api.themoviedb.org/3/movie/popular?language=en-US&page={j}"

    headers = {
        "accept": "application/json",
        "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiI4MGI4MTg5ZmYzZWQzNjljMDZhNzE0NTdiNGJlMmU5ZCIsIm5iZiI6MTc0MzA3NDM0NS4xMDMwMDAyLCJzdWIiOiI2N2U1MzQyOTMzYTc0MzQxZTMxMDlmNzUiLCJzY29wZXMiOlsiYXBpX3JlYWQiXSwidmVyc2lvbiI6MX0.MjLf9ZxV-09-rN9_9InskU5W2VCpBss7b5DR6dvUnwY"
    }

    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        print(f"Failed at page {j}: {response.status_code}")
        continue

    data = response.json()

    popularity = []
    vote_count = []
    original_language = []
    title = []
    vote_average = []
    overview = []
    release_data = []

    for i in data['results']:
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_language.append(i['original_language'])
        title.append(i['title'])
        vote_average.append(i['vote_average'])  # ✅ corrected
        overview.append(i['overview'])
        release_data.append(i['release_date'])

    d = {
        'title': title,
        "overview": overview,
        "release_data": release_data,
        'popularity': popularity,
        'vote_count': vote_count,
        'original_language': original_language,
        'vote_average': vote_average,
    }

    min_len = min(len(v) for v in d.values())
    d_trimmed = {k: v[:min_len] for k, v in d.items()}
    df = pd.DataFrame(d_trimmed)

    all_dfs.append(df)

# Concatenate all DataFrames into a final one
final = pd.concat(all_dfs, ignore_index=True)

# Optionally save to CSV
# final.to_csv("tmdb_popular_movies.csv", index=False)
```
