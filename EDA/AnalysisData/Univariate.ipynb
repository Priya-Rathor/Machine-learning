# Univariate Analysis

The analysis which is going to do independelly by  an colum , call Univariate Analysis

# type of data 
 * Numerical data
 * categorical  data 

```python
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# Categorical data

a. Countplot

```python
sns.countplot(df['Survived'])
```

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(df['Pclass'])
```

# b. Piechart

```python
df['Survived'].value_counts().plot(kind='pie')
```

```python
df['Sex'].value_counts().plot(kind='pie', autopct = '%.2f')
```

# 2. Numerical Data

# a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'])
```

```python
plt.hist(df['Age'], bins=60)
```

```python
sns.distplot(df['Age'])
```

# c. Boxplot

```python
sns.boxplot(df['Age'])
```
